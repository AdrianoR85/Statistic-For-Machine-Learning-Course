# Básico Sobre Probabilidade

A probabilidade é um ramo da matemática que estuda a chance de um evento ocorrer. Ela é expressa como um número entre 0 e 1, onde 0 significa que o evento é impossível e 1 significa que o evento é certo. Quanto maior a probabilidade, maior a chance de o evento acontecer.

**Conceitos Chave:**

*   **Experimento Aleatório:** Um processo cujo resultado não pode ser previsto com certeza (ex: jogar um dado).
*   **Espaço Amostral (Ω):** O conjunto de todos os resultados possíveis de um experimento aleatório (ex: ao jogar um dado, Ω = {1, 2, 3, 4, 5, 6}).
*   **Evento:** Um subconjunto do espaço amostral (ex: o evento 'sair um número par' em um dado é {2, 4, 6}).

**Fórmula Básica da Probabilidade:**

```
P(A) = (Número de resultados favoráveis ao evento A) / (Número total de resultados possíveis no espaço amostral)
```

**Exemplo:**

Qual a probabilidade de tirar um número par ao jogar um dado justo de 6 faces?

*   Resultados favoráveis (números pares): {2, 4, 6} -> 3 resultados
*   Total de resultados possíveis: {1, 2, 3, 4, 5, 6} -> 6 resultados

```
P(número par) = 3 / 6 = 1/2 = 0.5 ou 50%
```

## Permutação

A **permutação** é um arranjo de todos ou parte de um conjunto de itens onde a ordem dos itens importa. Em outras palavras, é o número de maneiras de organizar um conjunto de objetos em uma sequência ou ordem.

Existem dois tipos principais de permutação:

1.  **Permutação Simples (sem repetição):** Quando todos os itens são distintos e queremos arranjá-los.

    A fórmula para a permutação de *n* objetos tomados *k* a *k* é:
    $$P(n, k) = \frac{n!}{(n-k)!}$$

    Se *k = n* (ou seja, estamos arranjando todos os *n* objetos), a fórmula simplifica para:
    $$P(n, n) = n!$$

    *Onde `n!` (n fatorial) é o produto de todos os inteiros positivos de 1 a n (ex: `5! = 5 * 4 * 3 * 2 * 1 = 120`).*

2.  **Permutação com Repetição:** Quando os itens podem se repetir.

    A fórmula para a permutação de *n* objetos onde há $n_1$ repetições do tipo 1, $n_2$ repetições do tipo 2, ..., $n_k$ repetições do tipo k é:
    $$P_r = \frac{n!}{n_1! n_2! ... n_k!}$$

### Exemplos no notebook:

*   `math.factorial(3)` calcula $3! = 3 \times 2 \times 1 = 6$. Isso representa o número de maneiras de organizar 3 itens distintos.

*   `math.factorial(36) / math.factorial(36-5)` calcula $P(36, 5) = \frac{36!}{(36-5)!} = \frac{36!}{31!}$. Isso representa o número de permutações de 36 itens, pegando 5 de cada vez, onde a ordem importa.

## Combinação

A **combinação** é uma maneira de selecionar itens de um conjunto onde a ordem dos itens selecionados *não importa*. Ao contrário da permutação, onde a ordem é crucial, na combinação, um grupo de itens é considerado o mesmo, independentemente da sequência em que foram escolhidos.

Existem dois tipos principais de combinação:

1.  **Combinação Simples (sem repetição):** Quando os itens são distintos e não podemos selecioná-los mais de uma vez.

    A fórmula para a combinação de *n* objetos tomados *k* a *k* é:
    $$C(n, k) = \binom{n}{k} = \frac{n!}{k!(n-k)!}$$

    *Onde `n!` (n fatorial) é o produto de todos os inteiros positivos de 1 a n.* A notação $\binom{n}{k}$ é lida como "n escolhe k".

2.  **Combinação com Repetição:** Quando os itens podem ser selecionados mais de uma vez.

    A fórmula para a combinação de *n* objetos tomados *k* a *k* com repetição é:
    $$C_r(n, k) = \binom{n+k-1}{k} = \frac{(n+k-1)!}{k!(n-1)!}$$

### Exemplos no notebook:

*   `math.factorial(6) / (math.factorial(2) * math.factorial(6-2))` calcula $C(6, 2) = \frac{6!}{2!(6-2)!} = \frac{6!}{2!4!} = 15$. Isso representa o número de maneiras de escolher 2 itens de um conjunto de 6 itens distintos, sem se importar com a ordem.

*   `math.factorial(6 + 2 - 1) / (math.factorial(2) * math.factorial(6-1))` calcula $C_r(6, 2) = \frac{(6+2-1)!}{2!(6-1)!} = \frac{7!}{2!5!} = 21$. Isso representa o número de maneiras de escolher 2 itens de um conjunto de 6 itens com repetição.

````python
math.factorial(6) / (math.factorial(2) * math.factorial(6-2))
````

````python
math.factorial(6 + 2 - 1) / (math.factorial(2) * math.factorial(6-1))
````

## Teoria dos Conjuntos

A Teoria dos Conjuntos é fundamental em probabilidade porque ela fornece a linguagem e as ferramentas para descrever eventos e espaços amostrais. Em probabilidade:

*   **Espaço Amostral (Ω):** É o conjunto de todos os resultados possíveis de um experimento aleatório. Por exemplo, ao lançar um dado, o espaço amostral é Ω = {1, 2, 3, 4, 5, 6}.
*   **Evento:** É qualquer subconjunto do espaço amostral. Por exemplo, o evento 'obter um número par' é o conjunto A = {2, 4, 6}.

As operações de conjunto – **interseção**, **união** e **diferença** – têm significados diretos em probabilidade:

*   **Interseção (A ∩ B):** Representa o evento em que *ambos* os eventos A e B ocorrem simultaneamente. Se A = {2, 4, 6} e B = {1, 2, 3}, A ∩ B = {2} (ou seja, o resultado é 2).
*   **União (A ∪ B):** Representa o evento em que *pelo menos um* dos eventos A ou B ocorre. Usando os mesmos A e B, A ∪ B = {1, 2, 3, 4, 6}.
*   **Diferença (A \ B):** Representa o evento em que A ocorre *mas* B não ocorre. Se A = {2, 4, 6} e B = {1, 2, 3}, A \ B = {4, 6}.

Essas ferramentas nos permitem combinar e analisar eventos complexos, calculando suas probabilidades de ocorrência.

### Interseção

````python
a = [0,1,2,3,4,5,6,7]
b = [0,2,4,6,8]
````

````python
set(a) and set(b)
````

### União

````python
set(a) | set(b)
````

### Diferença

````python
set(a).difference(set(b))
````

````python
set(b).difference(set(a))
````
