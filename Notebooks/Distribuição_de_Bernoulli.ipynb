
# Bernoulli


## Conceito



A Distribuição de Bernoulli é uma distribuição de probabilidade discreta que modela um experimento aleatório com apenas dois resultados possíveis: sucesso (com probabilidade 'p') ou fracasso (com probabilidade '1-p'). É a base para muitas outras distribuições de probabilidade.


## Código

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
dados_bernoulli = bernoulli.rvs(size=1000, p=0.8)
```

```python
dados_bernoulli
```

```python
np.unique(dados_bernoulli, return_counts=True)
```

```python
sns.displot(dados_bernoulli)
plt.show()
```
