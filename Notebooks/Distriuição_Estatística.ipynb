## Variávies Contínuas

### Distribuição normal - Gaussian distribution

```python
dados_normal = stats.norm.rvs(size = 1000, random_state=1)
```

```python
min(dados_normal), max(dados_normal)
```

```python
sns.displot(dados_normal, kde=True)
```

```python
dados_normal.mean(), np.median(dados_normal), stats.mode(dados_normal), np.var(dados_normal), np.std(dados_normal)
```

```python
# Entre a média e a média + std
np.sum( ( (dados_normal >= np.mean(dados_normal)) & (dados_normal <= np.mean(dados_normal) + np.std(dados_normal)) ) )
```

```python
# Entre a média e a média - std
np.sum( ( (dados_normal >= np.mean(dados_normal) - np.std(dados_normal)) & (dados_normal <= np.mean(dados_normal)) ) )
```

```python
# Entre a média - std e a média + std
np.sum( ( (dados_normal >= np.mean(dados_normal) - np.std(dados_normal)) \
         & (dados_normal <= np.mean(dados_normal) + np.std(dados_normal)) ) )
```

### Distribuição gamma



*   Distribuição geral com valores assimétricos à direita
*   Análise de tempo de vida de produtos
*   Inadiplência com valores agregados
*   Quantidade de chuva acumulada em um resevatório
*   Tempo de reação de um motorista de acordo com a idade



```python
from  scipy.stats import gamma
```

```python
dados_gamma =  gamma.rvs(a = 5, size = 1000)
```

```python
sns.displot(dados_gamma, kde=True)
```

```python
dados_uniforme = uniform.rvs(size=1000)
```
