# Teste de Machine Learning

## Preparando os dados

```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Estatística para Ciência de Dados e Machine Learning/Bases de dados/credit_data.csv')
```

```python
dataset.dropna(inplace=True)
dataset.shape
```

```python
dataset
```

```python
X = dataset.iloc[:,1:4].values
Y = dataset.iloc[:,4].values
```

```python
X
```

```python
Y
```

## Importando Bibliotecas para o teste

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

## Treinamento


```python
from numpy.random import logistic
results_naive_bayes = []
results_logistics = []
results_forest = []

for i in range(30):
  x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=i)

  naive_bayes = GaussianNB()
  naive_bayes.fit(x_train, y_train)
  forecast_nb = naive_bayes.predict(x_test)
  results_naive_bayes.append(accuracy_score(y_test, forecast_nb))

  logistics_regression = LogisticRegression()
  logistics_regression.fit(x_train, y_train)
  forecast_lr = logistics_regression.predict(x_test)
  results_logistics.append(accuracy_score(y_test, forecast_lr))

  forest = RandomForestClassifier()
  forest.fit(x_train, y_train)
  forecast_f = forest.predict((x_test))
  results_forest.append(accuracy_score(y_test, forecast_f))
```

## Resultados

```python
print(results_naive_bayes)
```

```python
print(results_logistics)
```

```python
print(results_forest)
```

## Transformando os dados para uma nparray


```python
results_naive_bayes = np.array(results_naive_bayes)
results_logistics = np.array(results_logistics)
results_forest = np.array(results_forest)
```

## Validação Cruzada

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
results_naive_bayes_cv = []
results_logistics_cv = []
results_forest_cv = []

for i in range(30):
  kfold = KFold(n_splits=10, shuffle=True, random_state=i)

  naive_bayes = GaussianNB()
  scores_nb = cross_val_score(naive_bayes, X, Y, cv=kfold)
  results_naive_bayes_cv.append(scores_nb.mean())

  logistics_regression = LogisticRegression()
  scores_lr = cross_val_score(logistics_regression, X, Y, cv=kfold)
  results_logistics_cv.append(scores_lr.mean())

  forest = RandomForestClassifier()
  scores_f = cross_val_score(forest, X, Y, cv=kfold)
  results_forest_cv.append(scores_f.mean())
```

### Resultado

```python
print(f"Naive Bayes Results: {np.array(results_naive_bayes_cv)}\n")
print(f"Logistic Regression Results: {np.array(results_logistics_cv)}\n")
print(f"Random Forest: {np.array(results_forest_cv)}")
```
