# Poisson

A **Distribuição de Poisson** é uma distribuição de probabilidade que nos ajuda a prever a probabilidade de um certo número de eventos ocorrerem em um intervalo fixo de tempo ou espaço, se esses eventos acontecem com uma taxa média conhecida e independentemente do tempo desde o último evento.

### **Conceito:**
Imagine que você está observando um fenômeno que ocorre de forma aleatória e em intervalos de tempo ou espaço. A Distribuição de Poisson calcula a chance de você observar `x` ocorrências desse fenômeno dentro de um período ou área específica, sabendo apenas a média de quantas vezes esse fenômeno costuma acontecer.

### **Explicação:**
Os eventos que seguem uma distribuição de Poisson têm algumas características importantes:

1.  **Eventos Raros:** Geralmente, a distribuição é usada para modelar eventos que são relativamente raros dentro do intervalo observado.
2.  **Independência:** A ocorrência de um evento não afeta a probabilidade de outro evento ocorrer.
3.  **Taxa Média Constante (λ - lambda):** A taxa média de ocorrência de eventos é constante durante todo o intervalo. Esse valor médio é representado pela letra grega lambda (λ).
4.  **Eventos Discretos:** Contamos o número de ocorrências (0, 1, 2, 3, ...), então é uma distribuição discreta.

A fórmula da probabilidade de `x` eventos ocorrerem é:

$P(X=x) = \frac{\lambda^x \cdot e^{-\lambda}}{x!}$

Onde:
*   $P(X=x)$ é a probabilidade de `x` eventos ocorrerem.
*   $\lambda$ (lambda) é a taxa média de ocorrência de eventos por intervalo.
*   $e$ é a base do logaritmo natural (aproximadamente 2.71828).
*   $x!$ é o fatorial de `x` (o produto de todos os inteiros positivos até `x`).

### **Quando usar?**
Você deve usar a Distribuição de Poisson quando quiser modelar o número de vezes que um evento ocorre em um determinado intervalo contínuo (tempo, distância, área ou volume), e as seguintes condições são verdadeiras:

*   Você está contando o número de ocorrências de um evento, não a sua intensidade.
*   Os eventos ocorrem independentemente uns dos outros.
*   A taxa média de ocorrência dos eventos é constante no intervalo.
*   A probabilidade de dois ou mais eventos ocorrerem exatamente no mesmo ponto (ou momento) é desprezível.

**Exemplos Práticos:**

1.  **Chamadas Telefônicas:** Número de chamadas que um call center recebe em uma hora.
2.  **Acidentes:** Número de acidentes de carro em uma determinada estrada por mês.
3.  **Falhas de Equipamento:** Número de falhas em uma máquina em uma semana.
4.  **Bugs de Software:** Número de bugs encontrados em um software recém-lançado por dia.
5.  **Partículas Alfa:** Número de partículas alfa emitidas por uma fonte radioativa em um período de tempo.

## Código

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python

# Definir a taxa média (lambda) de ocorrência de eventos
# Exemplo: Em média, 3 clientes chegam por hora em uma loja.
lambda_val = 3

# Calcular a probabilidade de x eventos ocorrerem
# Qual a probabilidade de 0 clientes chegarem em uma hora?
prob_0_events = poisson.pmf(0, lambda_val)
print(f"Probabilidade de 0 eventos: {prob_0_events:.4f}")

# Qual a probabilidade de 3 clientes chegarem em uma hora (exatamente a média)?
prob_3_events = poisson.pmf(3, lambda_val)
print(f"Probabilidade de 3 eventos: {prob_3_events:.4f}")

# Qual a probabilidade de 5 clientes chegarem em uma hora?
prob_5_events = poisson.pmf(5, lambda_val)
print(f"Probabilidade de 5 eventos: {prob_5_events:.4f}")

# Calcular a probabilidade cumulativa (P(X <= x))
# Qual a probabilidade de até 2 clientes chegarem em uma hora (0, 1 ou 2)?
prob_le_2_events = poisson.cdf(2, lambda_val)
print(f"Probabilidade de até 2 eventos: {prob_le_2_events:.4f}")

# Visualizar a Distribuição de Poisson

x_values = np.arange(0, 10) # Número de eventos de 0 a 9
probabilities = [poisson.pmf(x, lambda_val) for x in x_values]

plt.figure(figsize=(10, 6))
sns.barplot(x=x_values, y=probabilities, color='skyblue')
plt.title(f'Distribuição de Poisson (λ = {lambda_val})')
plt.xlabel('Número de Eventos (x)')
plt.ylabel('Probabilidade P(X=x)')
plt.xticks(x_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### **Explicação do Código de Exemplo:**

Ele demonstra como usar a Distribuição de Poisson com a biblioteca `scipy.stats` em Python.

1.  **`lambda_val = 3`**: Definimos a taxa média de ocorrência (`λ`) como 3. Isso pode representar, por exemplo, a média de clientes que chegam em uma loja por hora.
2.  **`poisson.pmf(x, lambda_val)`**: Calcula a **Função Massa de Probabilidade (PMF)**, que nos dá a probabilidade de ocorrer *exatamente* `x` eventos. No exemplo, calculamos a probabilidade de 0, 3 e 5 clientes.
3.  **`poisson.cdf(x, lambda_val)`**: Calcula a **Função de Distribuição Cumulativa (CDF)**, que nos dá a probabilidade de ocorrer *até* `x` eventos (ou seja, `P(X <= x)`). No exemplo, calculamos a probabilidade de chegarem 0, 1 ou 2 clientes.
4.  **Visualização**: Um gráfico de barras é gerado para mostrar a distribuição das probabilidades para diferentes números de eventos, dado o `λ` definido. Você pode ver que a maior probabilidade está em torno da média (3 eventos).
