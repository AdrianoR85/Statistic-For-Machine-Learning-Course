```python
import pandas as pd
import math
import statistics
from scipy.stats.mstats import gmean, hmean

```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Estatística para Ciência de Dados e Machine Learning/Bases de dados/census.csv')
```

```python
print(f"Mean: {dataset['age'].mean()}")
```

```python
print(f"Harmonic Mean: {hmean(dataset['age'])}")
```

```python
print(f"Geometric Mean: {gmean(dataset['age'])}")
```

```python
def quandratic_mean(data):
  return math.sqrt(sum([x**2 for x in data])/len(data))
```

```python
print(f"Quandrantic Mean: {quandratic_mean(dataset['age'])}")
```

```python
print(f"Median: {dataset['age'].median()}")
```

```python
print(f"Mode: {statistics.mode(dataset['age'])}")
```
